```python
# ============================================
# SIMULADOR DE IA PARA ILUSTRADORES
# Lista de diccionarios
# Nivel: Preparatoria
# ============================================

# Lista vacía
ilustradores = []

# Diccionario vacío
losilustradores = {}

print("===================================")
print("   IA DE ILUSTRADORES DIGITALES")
print("===================================")

while True:

    print("\n========= MENÚ =========")
    print("1. Agregar ilustrador")
    print("2. Mostrar ilustradores")
    print("3. Buscar ilustrador")
    print("4. Modificar ilustrador")
    print("5. Eliminar ilustrador")
    print("6. Simulación IA")
    print("7. Salir")

    opcion = input("Selecciona una opción: ")

    # ============================================
    # AGREGAR
    # ============================================
    if opcion == "1":

        nombre = input("Nombre del ilustrador: ")
        estilo = input("Estilo de dibujo: ")
        programa = input("Programa que utiliza: ")
        nivel = input("Nivel (Principiante/Intermedio/Pro): ")

        # Diccionario
        losilustradores = {
            "nombre": nombre,
            "estilo": estilo,
            "programa": programa,
            "nivel": nivel
        }

        # Guardar en lista
        ilustradores.append(losilustradores)

        print("\nIlustrador agregado correctamente.")

    # ============================================
    # MOSTRAR
    # ============================================
    elif opcion == "2":

        if len(ilustradores) == 0:
            print("\nLa lista está vacía.")
        else:
            print("\n===== LISTA DE ILUSTRADORES =====")

            contador = 1

            for datos in ilustradores:

                print("\nIlustrador", contador)
                print("Nombre:", datos["nombre"])
                print("Estilo:", datos["estilo"])
                print("Programa:", datos["programa"])
                print("Nivel:", datos["nivel"])

                contador = contador + 1

    # ============================================
    # BUSCAR
    # ============================================
    elif opcion == "3":

        buscar = input("Escribe el nombre a buscar: ")

        encontrado = False

        for datos in ilustradores:

            if datos["nombre"] == buscar:

                print("\nIlustrador encontrado")
                print("Nombre:", datos["nombre"])
                print("Estilo:", datos["estilo"])
                print("Programa:", datos["programa"])
                print("Nivel:", datos["nivel"])

                encontrado = True

        if encontrado == False:
            print("\nNo se encontró el ilustrador.")

    # ============================================
    # MODIFICAR
    # ============================================
    elif opcion == "4":

        modificar = input("Nombre del ilustrador a modificar: ")

        encontrado = False

        for datos in ilustradores:

            if datos["nombre"] == modificar:

                print("\nNuevo estilo:")
                datos["estilo"] = input()

                print("Nuevo programa:")
                datos["programa"] = input()

                print("Nuevo nivel:")
                datos["nivel"] = input()

                print("\nDatos modificados correctamente.")

                encontrado = True

        if encontrado == False:
            print("\nNo existe ese ilustrador.")

    # ============================================
    # ELIMINAR
    # ============================================
    elif opcion == "5":

        eliminar = input("Nombre del ilustrador a eliminar: ")

        encontrado = False

        for datos in ilustradores:

            if datos["nombre"] == eliminar:

                ilustradores.remove(datos)

                print("\nIlustrador eliminado.")

                encontrado = True

        if encontrado == False:
            print("\nNo se encontró el ilustrador.")

    # ============================================
    # SIMULACIÓN IA
    # ============================================
    elif opcion == "6":

        print("\n===== IA DE DIBUJO =====")

        tipo = input("¿Qué deseas dibujar? ")

        if tipo == "anime":
            print("IA: Te recomiendo usar colores brillantes y sombras suaves.")

        elif tipo == "realista":
            print("IA: Usa detalles en ojos, piel y luces para más realismo.")

        elif tipo == "pixel art":
            print("IA: Utiliza pocos colores y cuadros pequeños.")

        else:
            print("IA: Estilo creativo detectado. ¡Sigue practicando!")

    # ============================================
    # SALIR
    # ============================================
    elif opcion == "7":

        print("\nGracias por usar el sistema IA.")
        break

    # ============================================
    # OPCIÓN INVÁLIDA
    # ============================================
    else:
        print("\nOpción incorrecta.")
```
